```python
##Step 1: Merge order_details with pizzas on 'pizza_id'
merged1 = pd.merge(order_details, pizzas, on='pizza_id', how='left')

# Step 2: Merge the result with pizza_types on 'pizza_type_id'
merged2 = pd.merge(merged1, pizza_types, on='pizza_type_id', how='left')

# Step 3: Merge with orders on 'order_id'
final_merged_dataset = pd.merge(merged2, orders, on='order_id', how='left')

# Save the final merged dataset to a new CSV file
final_merged_dataset.to_csv("merged_pizza_dataset.csv", index=False)
```

##  Convert date/time for time-based analysi

```python
dataset.info()
```

```python
dataset["date"] = pd.to_datetime(dataset["date"])
dataset['time'] = pd.to_datetime(dataset["time"]) ## converting data and time in object data type to datetime data type
dataset['hour'] = dataset['time'].dt.hour
dataset['weekday'] = dataset['date'].dt.day_name()
dataset['month'] = dataset['date'].dt.month_name()
```

```python
dataset
```

```python
dataset.info()
```


# add revenue column

```python
dataset['revenue'] = dataset['price'] * dataset['quantity']
```

## 1. Total revenue/sales

```python
total_revenue = dataset['revenue'].sum()
```

## 5 Average pizza price

```python
average_pizza_price = dataset[['pizza_id', 'price']].drop_duplicates()['price'].mean()
```

# 6. Peak hours of sales

```python
peak_sales_hours = dataset.groupby('hour')['revenue'].sum().sort_values(ascending=False)
```

```python
peak_sales_hours
```

# 7. Total sales by weekday

```python
sales_by_weekday = dataset.groupby('weekday')['revenue'].sum().sort_values(ascending=False)
```

```python
sales_by_weekday
```

# 8. Top 5 bestselling pizzas

```python
top_5_pizzas = dataset.groupby('pizza_id')['quantity'].sum().sort_values(ascending=False).head(5)
```

```python
top_5_pizzas
```

# 9. Sales made in each month
